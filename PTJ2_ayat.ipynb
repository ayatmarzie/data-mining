```python
import pandas as pd
import numpy as np
df = pd.read_excel('Absenteeism_at_work_Project.xlsx')
df['Body mass index'] = df.apply(
    lambda row: (row['Weight']/((row['Height']/100)*(row['Height']/100))) if np.isnan(row['Body mass index']) else row['Body mass index'],
    axis=1
    
)
df['Body mass index']=round(df['Body mass index'])
```

```python
df['Height'] = df.apply(
    lambda row: np.sqrt(row['Weight']*10000/row['Body mass index']) if np.isnan(row['Height']) else row['Height'],
    axis=1
    
)
df['Height']=round(df['Height'])
```

```python
df['Weight'] = df.apply(
    lambda row: (row['Body mass index']*((row['Height']/100)*(row['Height']/100))) if np.isnan(row['Weight']) else row['Weight'],
    axis=1
    
)
df['Weight']=round(df['Weight'])
```

```python
df['Social drinker'] = df['Social drinker'].fillna(df.groupby('ID')['Social drinker'].transform('first'))
df['Social smoker'] = df['Social smoker'].fillna(df.groupby('ID')['Social smoker'].transform('first'))
df['Pet'] = df['Pet'].fillna(df.groupby('ID')['Pet'].transform('first'))
df['Son'] = df['Son'].fillna(df.groupby('ID')['Son'].transform('first'))
df['Education'] = df['Education'].fillna(df.groupby('ID')['Education'].transform('first'))
df['Disciplinary failure'] = df['Disciplinary failure'].fillna(df.groupby('ID')['Disciplinary failure'].transform('first'))
df['Work load Average/day '] = df['Work load Average/day '].fillna(df.groupby('ID')['Work load Average/day '].transform('first'))
df['Age'] = df['Age'].fillna(df.groupby('ID')['Age'].transform('first'))
df['Service time'] = df['Service time'].fillna(df.groupby('ID')['Service time'].transform('first'))
df['Distance from Residence to Work'] = df['Distance from Residence to Work'].fillna(df.groupby('ID')['Distance from Residence to Work'].transform('first'))
df['Transportation expense'] = df['Transportation expense'].fillna(df.groupby('ID')['Transportation expense'].transform('first'))
df['Height'] = df['Height'].fillna(df.groupby('ID')['Height'].transform('first'))
df['Body mass index'] = df['Body mass index'].fillna(df.groupby('ID')['Body mass index'].transform('first'))
```

```python
df["Hit target"].fillna( method ='ffill', inplace = True)
df["Month of absence"].fillna( method ='ffill', inplace = True)
```

```python
for column in ['Reason for absence', 'Absenteeism time in hours',]:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
import scipy.spatial
from collections import Counter
class KNN:
    def __init__(self, k):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def distance(self, X1, X2):
        distance = scipy.spatial.distance.euclidean(X1, X2)

    def predict(self, X_test):
        final_output = []
        for i in range(len(X_test)):
            d = []
            votes = []
            for j in range(len(X_train)):
                dist = scipy.spatial.distance.euclidean(X_train[j] , X_test[i])
                d.append([dist, j])
            d.sort()
            d = d[0:self.k]
            for d, j in d:
                votes.append(y_train[j])
            ans = Counter(votes).most_common(1)[0][0]
            final_output.append(ans)

        return final_output

    def score(self, X_test, y_test):
        predictions = self.predict(X_test)
        return (predictions == y_test).sum() / len(y_test)
```
